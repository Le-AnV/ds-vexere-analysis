```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Đọc dữ liệu

```python
df = pd.read_csv("../../data/processed/sg_nt_11102025_cleaned_data.csv")
```

```python
print("Kích thước dữ liệu:", df.shape)
print("\n 5 dòng đầu:")
display(df.head())
```

```python
print("\n Thông tin chi tiết:")
df.info()
```

```python
print("\n Mô tả thống kê cơ bản:")
display(df.describe(include='all'))
```

# 2. Kiểm tra giá trị bị thiếu

```python
print("\n Kiểm tra giá trị bị thiếu:")
display(df.isnull().sum())
```

# 3. Biểu đồ phân bố cho các cột số

```python
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], kde=True, color='steelblue')
    plt.title(f"Phân bố giá trị của '{col}'")
    plt.show()
```

# 4. Biểu đồ tần suất cho các cột dạng phân loại

```python
cat_cols = df.select_dtypes(exclude=np.number).columns
for col in cat_cols:
    plt.figure(figsize=(7,3))
    df[col].value_counts().head(10).plot(kind='bar', color='teal')
    plt.title(f"Tần suất top giá trị trong '{col}'")
    plt.xlabel(col)
    plt.ylabel("Số lượng")
    plt.show()
```

# 5. Ma trận tương quan giữa các biến số

```python
if len(num_cols) > 1:
    plt.figure(figsize=(8,5))
    corr = df[num_cols].corr()
    sns.heatmap(corr, annot=True, cmap="Blues")
    plt.title("Ma trận tương quan giữa các biến số")
    plt.show()
```

# 6. Top 10 nhà xe có điểm đánh giá trung bình cao nhất

```python
# Chuyển cột bus_star sang kiểu số
df['bus_star'] = df['bus_star'].astype(str).str.extract('(\d+\.\d+)').astype(float)
top_rating = df.groupby('bus_company')['bus_star'].mean().sort_values(ascending=False).head(10)
print(top_rating)
sns.barplot(x=top_rating.values, y=top_rating.index)
plt.title("Top 10 Nhà xe có điểm đánh giá trung bình cao nhất")
plt.show()
```

# 7. Mối quan hệ giữa Giá vé và Chất lượng dịch vụ

```python
sns.scatterplot(data=df, x='price_discount', y='Chất lượng dịch vụ')
plt.title('Mối quan hệ giữa Giá vé và Chất lượng dịch vụ')
plt.show()
```

```python
# Lấy top 10 hãng xe phổ biến nhất
top_10_companies = df['bus_company'].value_counts().nlargest(10).index
df_top_10 = df[df['bus_company'].isin(top_10_companies)]

plt.figure(figsize=(12, 8))
sns.boxplot(data=df_top_10, x='price_discount', y='bus_company')
plt.title('Phân bố giá vé của 10 hãng xe phổ biến nhất')
plt.show()
```
