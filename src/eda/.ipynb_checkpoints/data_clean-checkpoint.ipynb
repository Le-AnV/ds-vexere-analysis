```python
import pandas as pd
import numpy as np
```

# 1. Đọc dữ liệu

```python
df_sgnt = pd.read_csv("../../data/raw/sg_nt_11102025.csv")
```

```python
df_sgnt.head()
```

```python
df_sgnt.columns
```

```python
df_sgnt.isnull().sum()*100/len(df_sgnt)
```

# 2. Xóa cột không cần thiết

```python
if 'Unnamed: 0' in df_sgnt.columns:
    df_sgnt = df_sgnt.drop('Unnamed: 0', axis=1)
if 'Unnamed: 20' in df_sgnt.columns:
    df_sgnt = df_sgnt.drop('Unnamed: 20', axis=1)
print("Đã xóa cột 'Unnamed: 0' và 'Unnamed: 20'.")
```

# 3. Làm sạch và chuyển đổi kiểu dữ liệu cho cột 'price_original'

```python
# 2.1. Xóa tất cả các ký tự không phải là số (dấu chấm, chữ,...)
df_sgnt['price_original'] = df_sgnt['price_original'].astype(str).str.replace(r'\D', '', regex=True)
df_sgnt['price_discount'] = df_sgnt['price_discount'].astype(str).str.replace(r'\D', '', regex=True)
# 2.2. Chuyển cột sang dạng số. Nếu có lỗi, giá trị đó sẽ thành NaN
df_sgnt['price_original'] = pd.to_numeric(df_sgnt['price_original'], errors='coerce')
df_sgnt['price_discount'] = pd.to_numeric(df_sgnt['price_discount'], errors='coerce')
print("Đã làm sạch và chuyển đổi cột 'price_original' và 'price_discount' sang dạng số.")
```

# 4. Liệt kê tất cả các cột cần điền giá trị

```python

rating_cols = [
    'An toàn', 'Thông tin chính xác', 'Thông tin đầy đủ',
    'Thái độ nhân viên', 'Tiện nghi & thoải mái', 'Chất lượng dịch vụ', 'Đúng giờ'
]

# 3.1 Lặp qua từng cột và điền giá trị thiếu
for col in rating_cols:
    # Điền giá trị thiếu bằng giá trị trung vị (Median) của từng nhóm 'bus_type'
    df_sgnt[col] = df_sgnt.groupby('bus_type')[col].transform(lambda x: x.fillna(x.median()))

    # Phòng trường hợp có một bus_type nào đó bị thiếu tất cả rating thì điền nốt bằng giá trị trung vị của toàn bộ cột
    if df_sgnt[col].isnull().any():
        # Loại bỏ cảnh báo (FutureWarning)
        df_sgnt[col] = df_sgnt[col].fillna(df_sgnt[col].median())

print("Đã điền các giá trị thiếu cho các cột đánh giá.")

# Kiểm tra kết quả
print("\nKiểm tra lại số lượng giá trị null sau khi làm sạch:")
print(df_sgnt.isnull().sum()*100/len(df_sgnt))

print("\nKiểm tra kiểu dữ liệu của các cột:")
print(df_sgnt.info())

print("\nKiểm tra số lượng hàng, cột:")
print(df_sgnt.shape)
```

# 5. Loại bỏ dữ liệu trùng lặp

```python
df_sgnt.drop_duplicates()
```

# 6. Thay đổi NaN thành ''

```python
columns_to_fill = ['bus_company', 'bus_star', 'bus_type', 'departure_time', 'pick_up_point', 'duration_h_m',
                   'arrival_time', 'drop_off_point', 'arrival_date', 'notification', 'discount_percent']
for col in columns_to_fill:
    df_sgnt[col] = df_sgnt[col].fillna(' ')
print(df_sgnt.head(20))
```

```python
df_sgnt.head(20)
```

# 7. Xuất file

```python
import os

# 1. Tách đường dẫn thư mục và tên file ra
output_dir = '../../data/processed'
file_name = 'sg_nt_11102025_cleaned_data.csv'
output_path = os.path.join(output_dir, file_name)

# 2. Kiểm tra xem thư mục đã tồn tại chưa, nếu chưa thì tạo nó
# os.makedirs sẽ tạo tất cả các thư mục trung gian nếu cần
# exist_ok=True sẽ không báo lỗi nếu thư mục đã tồn tại sẵn
os.makedirs(output_dir, exist_ok=True)

# 3. Lưu file
df_sgnt.to_csv(output_path, index=False, encoding='utf-8-sig')

print(f"Thư mục đã được tạo thành công.")
print(f"File đã được lưu thành công tại: {output_path}")
```
